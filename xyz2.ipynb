```python
import pandas as pd

# assuming 'data' is already a Python dict from your JSON

# Get the list of items
items = data.get("items", [])

# Flatten JSON into table
df = pd.json_normalize(items)

# Select useful columns
df = df[["id", "snippet.title", "snippet.assignable"]]
df = df.rename(columns={
    "id": "categoryId",
    "snippet.title": "category_name",
    "snippet.assignable": "assignable"
})

# Convert category_id to int
df["categoryId"] = pd.to_numeric(df["categoryId"], errors="coerce")

print(df.head())
```

```python
import pandas as pd

sf = pd.read_csv("new_IN_youtube_trending_data.csv", compression="zip", encoding="latin1", engine="python")

```

```python
sf.tail()
```

```python
sf["categoryId"] = pd.to_numeric(sf["categoryId"])
```

```python
merged = sf.merge(df, on="categoryId", how="left")

```

```python
merged
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(data=merged, y="category_name",
              order=merged["category_name"].value_counts().index)
plt.title("Trending Videos by Category")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=merged, x="view_count", y="likes", hue="category_name", alpha=0.6)
plt.xscale("log"); plt.yscale("log")   # log scale since numbers vary a lot
plt.title("Views vs Likes by Category")
plt.show()
```

```python
merged["like_ratio"] = merged["likes"] / (merged["dislikes"]+1)

plt.figure(figsize=(10,6))
sns.boxplot(data=merged, x="category_name", y="like_ratio")
plt.xticks(rotation=90)
plt.title("Like-to-Dislike Ratio by Category")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=merged, x="view_count", y="comment_count", hue="category_name", alpha=0.6)
plt.xscale("log"); plt.yscale("log")
plt.title("Comments vs Views by Category")
plt.show()
```

```python
merged["publishedAt"] = pd.to_datetime(merged["publishedAt"])
merged["publish_hour"] = merged["publishedAt"].dt.hour

plt.figure(figsize=(10,6))
sns.countplot(data=merged, x="publish_hour")
plt.title("Videos Published by Hour of Day")
plt.show()
```

```python
top_channels = merged["channelTitle"].value_counts().head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=top_channels.values, y=top_channels.index)
plt.title("Top 15 Trending Channels")
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(merged["tags"].astype(str))
wc = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Video Tags")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.histplot(merged['view_count'], bins=100, log_scale=True, color="skyblue")
plt.title("Distribution of Video Views (Log Scale)")
plt.xlabel("Views (log scale)")
plt.ylabel("Number of Videos")
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_cat = merged.groupby('category_name')['view_count'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_cat.values, y=top_cat.index, palette="viridis")
plt.title("Top 10 Categories by Total Views")
plt.xlabel("Total Views")
plt.ylabel("Category")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = merged[['view_count','likes','dislikes','comment_count','like_ratio']].corr()
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Heatmap of Engagement Metrics")
plt.show()
```
